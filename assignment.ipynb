### Importing Pymango library

```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient("mongodb://localhost:27017/")
```

```python
database = client['iNeuron']
```

```python
collection = database['Addmision_details']
```

```python

dataframe=pd.read_csv("https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv")
```

```python
dataframe = dataframe.to_dict('records') 
```

```python
records = collection.insert_many(dataframe)
```

```python
result = collection.find({})
```

```python
for i in result:
    print(i)
```

```python
query1 = {'University Rating' : 1}
answer = collection.find(query1)
for i in answer:
    print(i)
```

```python
query2 = {'CGPA' : {'$gt' : 9.5}}
answer = collection.find(query2)
for i in answer:
    print(i)

```

```python
query2 = {'CGPA' : {'$gt' : 9.5}}
answer = collection.find(query2).limit(2)
for ans in answer:
    print(ans)
```

```python
query2 = {'CGPA' : {'$gt' : 9.5}}
answer = collection.find(query2)
print(answer.count())
```
